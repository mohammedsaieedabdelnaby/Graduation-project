```python
from transformers import pipeline

# Initialize the Llama model
essay_generator = pipeline(
    "text-generation",
    model="meta-llama/Llama-3.2-3B-Instruct",  # Example model
    device_map="auto"  # Automatically uses GPU if available
)
```

```python
system_messages = {
    "Research Paper": "Please write this research paper with a clear structure, including an introduction that states the research question and significance, body paragraphs that present the methodology, findings, and analysis, and a conclusion that summarizes the key insights and implications. Use formal academic language, proper citations, and varied vocabulary while maintaining a cohesive flow throughout. Conclude with a strong summary that reinforces the main points and highlights the contribution to the field.",
    "Journalistic Essay": "Please write this journalistic essay with a clear structure, including an engaging introduction that captures attention, body paragraphs that present factual information, quotes, and perspectives, and a conclusion that reflects on the broader implications of the story. Use clear and concise language, varied vocabulary, and maintain an objective tone while weaving a compelling narrative. Conclude with a strong summary that reinforces the main points and encourages readers to reflect on the issue.",
    "Policy Paper": "Please write this policy paper with a clear structure, including an introduction that outlines the issue and its significance, body paragraphs that analyze current policies, present evidence, and propose recommendations, and a conclusion that emphasizes the importance of policy change. Use formal language, varied vocabulary, and maintain a logical flow throughout. Conclude with a strong summary that reinforces the main points and outlines the potential impact of the proposed policy.",
    "Persuasive Essay": "Please write this persuasive essay with a clear structure, including an introduction that presents a strong thesis statement, body paragraphs that provide compelling arguments supported by evidence and emotional appeals, and a conclusion that urges the reader to take action. Use varied vocabulary and maintain a cohesive flow throughout. Conclude with a strong summary that reinforces the main points and emphasizes the urgency of the call to action.",
    "Descriptive Essay": "Please write this descriptive essay with a clear structure, including an introduction that sets the scene or describes the subject, body paragraphs that evoke sensory details and vivid imagery, and a conclusion that reflects on the overall experience or significance of the description. Use varied vocabulary that captures the essence of the subject and engages the reader’s senses. Maintain a cohesive flow throughout the piece. Conclude with a strong summary that reinforces the main points and highlights the importance of the described experience.",
    "Professional Essay": "Please write this professional essay with a clear structure, including an introduction that outlines the topic's relevance in the workplace, body paragraphs that discuss key insights or best practices, and a conclusion that emphasizes actionable takeaways. Use formal and concise language, varied vocabulary, and maintain a cohesive flow throughout. Ensure clarity and professionalism in tone. Conclude with a strong summary that reinforces the main points and their implications in a professional context.",
    "Academic Essay": "Please write this academic essay with a clear structure, including an introduction that presents the thesis statement, body paragraphs that develop and support the main arguments with scholarly evidence, and a conclusion that summarizes the findings. Use formal language, varied vocabulary, and maintain a cohesive flow throughout. Ensure proper citations and a reference list where applicable. Conclude with a strong summary that reinforces the main points and reflects on their significance in the academic field.",
    "Admission Essay": "Please write this admission or scholarship essay with a clear structure, including an introduction that presents your personal background and aspirations, body paragraphs that highlight your achievements, experiences, and motivations, and a conclusion that reinforces your fit for the program or scholarship. Use a personal and authentic tone, varied vocabulary, and maintain a cohesive flow throughout. Conclude with a strong summary that emphasizes your goals and the impact of the opportunity on your future.",
    "Cover Letter": "Please write this application or cover letter with a clear structure, including an introduction that expresses your interest in the position or opportunity, body paragraphs that outline your relevant skills and experiences, and a conclusion that reiterates your enthusiasm and invites further discussion. Use professional and concise language, varied vocabulary, and maintain a clear and persuasive tone. Conclude with a strong summary that reinforces your suitability for the role and expresses gratitude for the consideration.",
    "Scientific Paper": "Please write this scientific paper with a clear structure, including an introduction that outlines the research question and hypothesis, body paragraphs that detail the methodology, results, and discussion, and a conclusion that summarizes the findings and their implications. Use precise and technical language, proper citations, and varied vocabulary while maintaining a logical flow throughout. Conclude with a strong summary that reinforces the main points and suggests areas for further research."
}
```

```python
def generate_essay(essay_type, prompt):
    # Retrieve the corresponding system message
    system_message = system_messages[essay_type]

    # Combine system message and user prompt into a single string
    combined_input = f"{system_message}\n\n{prompt}"

    # First layer: Generate initial essay output
    response_layer_1 = essay_generator(combined_input, max_length=2000)
    output_layer_1 = response_layer_1[0]["generated_text"].strip()

    # Remove the system message from the first output
    trimmed_response = output_layer_1.replace(system_message, "").strip()

    # Default prompt for the second layer
    default_prompt = "Refine the following essay and improve its clarity and coherence without telling me the changes that U did on the essay and without giving me any Extra NOTES or any furthur explainations :"

    # Combine trimmed response with the default prompt for the second layer
    second_layer_input = f"{default_prompt}\n\n{trimmed_response}"

    # Second layer: Refine the initial output using the default prompt
    refined_response = essay_generator(second_layer_input, max_length=2000)
    output_layer_2 = refined_response[0]["generated_text"].strip()
    output_layer_2 = output_layer_2.replace(default_prompt, "").strip()
    return output_layer_2

essay_types = list(system_messages.keys())
with gr.Blocks(css="""
    .header { color: #4CAF50; font-size: 24px; text-align: center; }
    .footer { text-align: center; color: #777; }
    .button { background-color: #4CAF50; color: white; border: none; padding: 10px 20px; text-align: center; text-decoration: none; display: inline-block; font-size: 16px; margin: 4px 2px; cursor: pointer; border-radius: 5px; }
    .textbox { border-radius: 5px; }
""") as demo:
    gr.Markdown("<h1 class='header'>Content Flow</h1>")
    gr.Markdown("<p style='text-align: center;'>Your AI-powered essay generator.</p>")

    with gr.Row():
        essay_type = gr.Dropdown(choices=essay_types, label="Choose Essay Type")
        user_input = gr.Textbox(label="Enter Essay Topic", placeholder="What would you like to write about?")

    generate_button = gr.Button("Generate Essay")

    output = gr.Textbox(label="Generated Essay", interactive=False, placeholder="Your generated essay will appear here...")

    # Connect button with function
    generate_button.click(generate_essay, inputs=[essay_type, user_input], outputs=[output])

    gr.Markdown("<hr>")
    gr.Markdown("<p class='footer'>Designed by Mohammed Saieed </p>")

# Launch the app
demo.launch()
```
